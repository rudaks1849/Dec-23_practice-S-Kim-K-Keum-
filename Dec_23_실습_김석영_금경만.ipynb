```python
path = "./data/Avengers.txt"

with open(path, "r", encoding="utf-8") as f:
    text = f.read()
```

```python
# 기호 제거
```

```python
cleaned = []

for ch in text:
    if ch.isalnum():
        cleaned.append(ch)
    else:
        cleaned.append(" ")
```

```python
# 단어 분리
```

```python
words = "".join(cleaned).split()
```

```python
# 각 단어별 빈도수를 딕셔너리 형태로 저장
```

```python
import re
from collections import Counter
```

```python
freq = dict(Counter(words))
```

```python
freq
```

```python
# 빈도 순 정렬 후 vocab 만들기
```

```python
items = list(freq.items())
```

```python
items
```

```python
def by_count(item):
    return item[1]
```

```python
items =  sorted(items, key=by_count, reverse=True)
```

```python
items
```

```python
vocab = {}
idx = 0
```

```python
for word, count in items:
    vocab[word] = idx
    idx += 1
```

```python
vocab
```

```python
list(vocab.items())[:20]
```

```python
# 텍스트를 입력받아 정수를 리턴하는 encoder 함수
```

```python
def encoder():
    sentence = input("단어를 입력하세요: ").lower()

    cleaned = []
    for ch in sentence:
        if ch.isalnum():
            cleaned.append(ch)
        else:
            cleaned.append(" ")

    words = "".join(cleaned).split()

    encoded = []
    for w in words:
        if w in vocab:
            encoded.append(vocab[w])

    print(encoded)
```

```python
encoder()
```
